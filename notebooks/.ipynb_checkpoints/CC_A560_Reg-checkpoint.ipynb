```python
data_loc = '/home/rdmtinez/Desktop/MScThesis/data_o/calibration/community_calibration/parsed_data/'
fname = 'calib_exp2-cell_counter_results.tsv'
```

```python
df= pd.read_csv(data_loc+fname, sep='\t')
df.head()
```

```python
dft = df.rename({'cc_sampling_time':'date_time', 
                        'type_of_cell':'cell_type',
                        'Sample_concentration':'concentration',
                'Filename':'well'}, axis=1).drop('Filename_full', axis=1)

dft.head()
dft['well'] = dft['well'].str[20:25]

dft['rep'] = dft['well'].str[3:]
dft['well'] = dft['well'].str[0:2]

dft = dft[['date_time', 'well', 'rep', 'concentration', 'cell_type']]

dft['date_time'] = pd.to_datetime(dft['date_time']).dt.round('11H')
```

```python
x2gb = dft.groupby(by='cell_type')

cell_counts_x2 = []
for name, group in x2gb:
    group.loc[group['well']=='F1', 'well'] = 'F11'
    group.loc[group['well']=='G1', 'well'] = 'G12'
    
    
    group.loc[group['well']=='F11', 'rep'] = ['01', '02', '03', '04']
    group.loc[group['well']=='G12', 'rep'] = ['01', '02', '03', '04']
    group.head()
    cell_counts_x2.append(group.drop(labels=['cell_type'], axis=1))

cell_counts_x2[0].head()
```

```python
# Mean & Standard Deviation Calculation and DF reformatting
x2_cell_counts = []
for df in cell_counts_x2:
    tdf = df.groupby(['well'], as_index=False).agg({'concentration':['mean', 'std'],'date_time':'last'})
    tdf.columns = ['well', 'conc_mean', 'conc_std', 'date_time']
    tdf = tdf.sort_values(by=['well'])
    tdf = tdf.set_index('well')
    tdf = tdf.drop('date_time',axis=1)
    x2_cell_counts.append(tdf)

    tdf
```

```python
calib_cc = x2_cell_counts.copy()
```

```python
B_c = calib_cc[0]
C_c = calib_cc[1]
```

```python

idx = B_c.index.values
idx
```
